```python
import matplotlib.pyplot as plt
```

```python
# 1행1열의 그래프 영역 생성 -> 리턴받는 ax객체를 활용하여 그래프를 그리게 됨
fig, ax = plt.subplots(1, 1)

# 생성된 그래프 영역(ax)에 리스트 데이터를 선 그래프로 표현.
# -> 리스트의 각 값은 y축이 되고, 리스트 값의 인덱스는 x축이 된다.
ax.plot(data)

# 그래프를 이미지로 저장하기 (필요한 경우만 사용, show()보다 먼저 호출되어야 함)
plt.savefig('first.png')

# 그래프 표시하기
plt.show()

# 그래프 관련 설정 해제
plt.close()
```

### 선 그래프 옵션 활용

```python
newborn = [436455, 435435, 438420, 406243, 357771]
year = [2013, 2014, 2015, 2016, 2017]
```

#### 선 그래프 표시하기

```python
# 1행1열의 그래프 영역 생성 
fig, ax = plt.subplots(1, 1)

# 리스트 데이터를 선 그래프로 표현.
# -> 리스트의 각 값은 y축이 되고, 리스트 값의 인덱스는 x축이 된다.
ax.plot(newborn, label='baby count', linestyle='--', marker='.', color='#ff6600')
plt.savefig('line1.png')

# plot()함수의 label속성을 그래프에 적용함 -> 범주
ax.legend()

plt.savefig('line2.png')

# 배경에 그리드 표시하기
ax.grid()
plt.savefig('line3.png')

# 그래프 제목, x축, y축 라벨 설정
ax.set_title('Newborn baby of year')
ax.set_xlabel('year')
ax.set_ylabel('newborn')
plt.savefig('line4.png')

# x축의 각 위치에 year 리스트의 값을 라벨로 적용
# -> 데이터 리스트의 '인덱스를 원소로 갖는 리스트'와
#    그 위치에 표시될 '텍스트를 원소로 갖는 리스트'를 설정
ax.set_xticks([0, 1, 2, 3, 4])
ax.set_xticklabels(year)
plt.savefig('line5.png')

# 그래프 표시하기
plt.show()

# 그래프 관련 설정 해제
plt.close()
```

### x, y축 범위 조절

```python
# 어느 학생의 학년별 성적 변화
grade = {"1year": 72, "2year": 83, "3year": 64, "4year": 92}
```

```python
# 그래프로 표시할 Key, Value 분리

# 딕셔너리의 value를 리스트로 추출
values = list(grade.values())
print(values)

# 딕셔너리의 key를 리스트로 추출
keys = list(grade.keys())
print(keys)
```

#### 그래프 표시하기

```python
# 1행1열의 그래프 영역 생성 
fig, ax = plt.subplots(1, 1)

# 리스트 데이터를 선 그래프로 표현.
# -> 리스트의 각 값은 y축이 되고, 리스트 값의 인덱스는 x축이 된다.
ax.plot(values, label='grade', color='#ff00ff')

ax.legend()                # plot()함수의 label속성을 그래프에 적용
ax.grid()                  # 배경에 그리드 표시하기
ax.set_title('Level Grade')    # 그래프 제목
ax.set_xlabel('level')         # x축 라벨
ax.set_ylabel('point')         # 점수

# x축의 좌표에 대한 리스트 생성
xpos = list(range(0, len(values)))

# x축의 각 위치에 year 리스트의 값을 라벨로 적용
# -> 데이터 리스트의 '인덱스를 원소로 갖는 리스트'와
#    그 위치에 표시될 '텍스트를 원소로 갖는 리스트'를 설정
ax.set_xticks(xpos)
ax.set_xticklabels(keys)
ax.set_ylim([0, 100])            # y축의 범위 설정
ax.set_xlim([0, 3])              # x축의 범위 설정

plt.show()
plt.close()                 # 그래프 설정 해제
```

#### x축을 좌표로 설정하기

```python
# 데이터 정의

newborn = [436455, 435435, 438420, 406243, 357771]
x = list(range(0, len(newborn)))
x
```

```python
# x축을 좌표값으로 갖는 막대 그래프 생성

# 가로 14inch, 세로 7inch의 1행 1열 그래프 영역 ax 준비
fig, ax = plt.subplots(1, 1, figsize=(14, 7))

# 막대 그래프 표현
ax.bar(x, newborn, label='신생아 수', color='#ecc2f9')

# 범주 표시하기
ax.legend()

# 기준축(x축) 라벨
ax.set_xlabel('년도')

# 데이터(y축) 라벨
ax.set_ylabel('신생아 수')

# 그래프 제목
ax.set_title('2013~2017년도 신생아 수', fontsize=24, color='#0066ff', pad=20)

# 격자 선 표시
ax.grid()

# 그래프의 x좌표에 표시될 텍스트 설정
# -> x축의 좌표를 의미하는 [0, 1, 2, 3, 4] 설정
ax.set_xticks(x)
# -> 지정된 좌표에 표시될 텍스트 리스트 설정
ax.set_xticklabels(['2013년', '2014년', '2015년', '2016년', '2017년'])

# 그래프의 데이터 수 만큼 반복하면서 텍스트 표시하기
for i, v in enumerate(newborn):
    # -> 표시할 텍스트
    str_val = "%d명" % v

    # -> x축 좌표, y축 좌표, 표시문구, 글자크기, 색상, 가로/세로 정렬
    ax.text(x[i], v, str_val, fontsize=13, color='#ff0000', horizontalalignment='center', verticalalignment='bottom')

# y축의 범위를 조절하여 전달하고자 하는 스토리를 극대화 하기
ax.set_ylim([300000, 450000])

plt.show()
plt.close() 
```

### 파이 그래프

```python
# 옵션으로 사용할 값 준비

# 표시할 데이터 설정
# -> 총 합이 100이 아닐 경우 라이브러리가 자동으로 비율을 계산함
ratio = [3700, 2800, 1300, 590, 600]
# 각 항목의 라벨
labels = ['Apple', '삼성', 'LG', 'Sony', '기타']
# 각 항목의 색상
colors = ['#ff6600', '#ff00ff', '#ffff00', '#00ffff', '#0066ff']
# 확대비율
explode = [0.0, 0.0, 0.0, 0.2, 0.1]
```

```python
# 그래프 구현

# 가로 10inch, 세로 10inch의 1행 1열 그래프 영역 ax 준비
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

ax.set_title('스마트폰 점유율', fontsize=24)

# 파이차트 표시(데이터,라벨,색상,확대,수치값 표시형식,그림자,시작각도)
# -> autopct 파라미터 설정 안할 경우 수치값 표시 안됨
#       의미: %0.2f (소수점 2째 자리까지 표시) + %% (순수한 %기호)
# -> startangle 기본값은 0도
# -> 각 데이터 항목들은 반시계 반향으로 회전하면서 배치됨
ax.pie(ratio, labels=labels, colors=colors, explode=explode,
           autopct='%0.2f%%', shadow=False, startangle=90)

#plt.savefig("pie.png", dpi=200, bbox_inches='tight')
plt.show()
plt.close()
```

### 산점도 그래프

```python
# 샘플데이터 준비

# 온도와 아이스크림 판매 수량
tmp = [ 23, 25, 26, 27, 28, 29, 30, 31, 33 ]
qty = [ 21, 23, 25, 28, 33, 35, 36, 32, 39 ]
```

```python
# 폰트 지정

# 기본 글자 크기
plt.rcParams["font.size"] = 15
```

```python
# 그래프 구현

# 가로 18inch, 세로 10inch의 1행 1열 그래프 영역 ax 준비
fig, ax = plt.subplots(1, 1, figsize=(18, 10))

# marker -> o, v, ^, <, >, 8, s, p, *, h, H, D, d, P, X
ax.scatter(tmp, qty, color='#ff6600', marker='o', label='판매수량')
ax.legend()
ax.grid()
ax.set_title('기온과 아이스크림 판매수량의 관계', fontsize=24, pad=12)
ax.set_ylabel('아이스크림 판매수량', fontsize=18, color='#0066ff')
ax.set_xlabel('기온', fontsize=18, color='#0066ff')

#plt.savefig("scatter.png", dpi=200, bbox_inches='tight')
plt.show()                                     # 그래프 표시하기
plt.close()   
```

### 서브 플롯

```python
# a행 b열의 그래프 영역 생성하기

fig, ax = plt.subplots(2, 3, figsize=(15, 10))
print(ax)
plt.show()
plt.close()
```

```python
# 서브플롯 영역안에 그래프 그리기

fig, ax = plt.subplots(1, 4, figsize=(20, 5))

ax[0].plot([0, 1, 2, 3, 4])
ax[1].bar(['a','b','c'], [10, 20, 30])
ax[2].pie([10, 20, 30])
ax[3].scatter([10, 20, 30, 40, 50], [1, 2, 4, 5, 6])

plt.show()
plt.close()
```

```python
# 2행 3열의 서브플롯 안에 개별적인 그래프 그리기

fig, ax = plt.subplots(2, 2, figsize=(20, 10))

ax[0][0].plot([0, 1, 2, 3, 4])
ax[0][1].bar(['a','b','c'], [10, 20, 30])
ax[1][0].pie([10, 20, 30])
ax[1][1].scatter([10, 20, 30, 40, 50], [1, 2, 4, 5, 6])

plt.savefig("sub3.png", dpi=200, bbox_inches='tight')
plt.show()
plt.close()
```
