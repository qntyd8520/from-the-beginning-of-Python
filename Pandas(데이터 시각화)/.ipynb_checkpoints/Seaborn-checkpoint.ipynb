```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# 그래프 전역설정
plt.rcParams["font.family"] = 'Malgun Gothic' # 한글 지원 폰트 설정
plt.rcParams["font.size"] = 18                # 기본 글자크기 설정
plt.rcParams["figure.figsize"] = (20, 10)     # 그래프 사이즈 (가로,세로)inch
plt.rcParams["axes.unicode_minus"] = False    # 유니코드 뺄셈기호 표시 사용 안함(깨짐방지)
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

### 단일 선 그래프 구현

```python
# 샘플 데이터
# 년도별 신생아 수 데이터
newborn = [436455, 435435, 438420, 406243, 357771]
# 년도
year = [2013, 2014, 2015, 2016, 2017]
# x축의 좌표로 사용할 0부터 1씩 증가하는 좌표값 리스트
newborn_x = list(range(0, len(newborn)))
```

#### 선 그래프 구현

```python
# 1행 1열의 그래픽 영역을 갖는 ax객체 생성
fig, ax = plt.subplots(1, 1)

# x축 좌표는 가급적 0부터 시작하는 좌표 형식으로 지정.
sns.lineplot(x=newborn_x, y=newborn, label='신생아수', ax=ax)

# 그래프 타이털 설정
ax.set_title('년도별 신생아 수 변동 추이', fontsize=24, pad=12)

# x축의 텍스트 설정
ax.set_xlabel('년도', fontsize=16, color='#555555')

# y축의 텍스트 설정
ax.set_ylabel('신생아 수', fontsize=16, color='#555555')

# x축 텍스트를 설정할 좌표 지정
ax.set_xticks(newborn_x)

# 지정된 좌표에 표시할 텍스트 지정
ax.set_xticklabels(year)

# 범주 표시
ax.legend()

# 배경 격자 표시
ax.grid()

# 이미지로 저장
#plt.savefig('seaborn1.png', dpi=200, bbox_inches='tight')

# 화면 출력
plt.show()

# 그래프 표시 종료
plt.close()
```

#### 다중 선 그래프 구현

```python
# 샘플 데이터 
# 2017년 1월~12월까지 도시별 교통사고 건수
seoul = [3166, 2728, 3098, 3172, 3284, 3247, 3268, 3308, 3488, 3312, 3375, 3179]
busan = [927, 857, 988, 955, 1014, 974, 1029, 1040, 1058, 971, 958, 982]
daegu = [933, 982, 1049, 1032, 1083, 1117, 1076, 1080, 1174, 1163, 1146, 1135]
inchun = [655, 586, 629, 669, 643, 627, 681, 657, 662, 606, 641, 663]
label = ['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월']
traffic_x = list( range( 0, len(label) ) )
```

```python
# 다중 선 그래프 출력

# 1행 1열의 그래픽 영역을 갖는 ax객체 생성
fig, ax = plt.subplots(1, 1)

# 총 4개의 선 그리기
sns.lineplot(x=traffic_x, y=seoul, label='서울', ax=ax)
sns.lineplot(x=traffic_x, y=busan, label='부산', ax=ax)
sns.lineplot(x=traffic_x, y=daegu, label='대구', ax=ax)
sns.lineplot(x=traffic_x, y=inchun, label='인천', ax=ax)

# 배경 격자 표시하기
ax.grid()

# 그래프 타이틀
ax.set_title('2017년 주요도시 교통사고', fontsize=24, pad=12)

# x축 텍스트
ax.set_xlabel('월', fontsize=16, color='#555555')

# y축 텍스트
ax.set_ylabel('교통사고', fontsize=16, color='#555555')

# x축의 범위 설정
ax.set_xlim([0, 11])

# y축의 범위 설정
ax.set_ylim([0, 4000])

# 범주 설정
ax.legend(title='도시', loc='center left')

# 텍스트를 출력할 x축 좌표 지정
ax.set_xticks(traffic_x)

# x축에 지정된 좌표위에 출력할 텍스트 지정
ax.set_xticklabels(label)

# 이미지로 저장
#plt.savefig('seaborn2.png', dpi=200, bbox_inches='tight')

# 그래프 표시하기
plt.show()

# 그래프 설정 종료
plt.close()
```

### 막대 그래프 구현

```python
# 셈플 데이터

values = [59, 23, 5, 38, 14, 16]
keys = ['아이언맨', '캡틴아메리카', '호크아이', '토르', '헐크', '블랙위도우']
```

```python
# 그래프 구현

# 1행 1열의 그래픽 영역을 갖는 ax객체 생성
fig, ax = plt.subplots(1, 1)

# 막대 그래프 구현하기
sns.barplot(x=keys, y=values, label='득표수', palette="Spectral", ax=ax)

# 범주 표시하기
ax.legend()

# x축 텍스트 표시
ax.set_xlabel('캐릭터', color='#ff0000', fontsize=14)

# y축 텍스트 표시
ax.set_ylabel('득표수', color='#0000ff', fontsize=14)

# 그래프 제목 구성
ax.set_title('어벤져스 인기투표', color='#ff6600', fontsize=28, pad=16)

# 배경 격자 구성
ax.grid()

# 이미지로 저장
#plt.savefig('seaborn3.png', dpi=200, bbox_inches='tight')

# 그래프 화면 출력
plt.show()

# 그래프 종료
plt.close()
```

#### 가로 막대그래프 

```python
# 1행 1열의 그래픽 영역을 갖는 ax객체 생성
fig, ax = plt.subplots(1, 1)

# 막대 그래프 구현하기
sns.barplot(x=values, y=keys, label='득표수', palette="Spectral", ax=ax)

# 범주 표시하기
ax.legend()

# y축 텍스트 표시
ax.set_ylabel('캐릭터', color='#ff0000', fontsize=14)

# x축 텍스트 표시
ax.set_xlabel('득표수', color='#0000ff', fontsize=14)

# 그래프 제목 구성
ax.set_title('어벤져스 인기투표', color='#ff6600', fontsize=28, pad=16)

# 배경 격자 구성
ax.grid()

# 이미지로 저장
plt.savefig('seaborn4.png', dpi=200, bbox_inches='tight')

# 그래프 화면 출력
plt.show()

# 그래프 종료
plt.close()
```

### 다중 막대 그래프

```python
# 샘플 데이터 

subject = ['국어', '국어', '영어', '영어', '수학', '수학', '과학', '과학']
season = ['중간고사', '기말고사', '중간고사', '기말고사', '중간고사', '기말고사', '중간고사', '기말고사']
point = [98, 82, 77, 93, 52, 88, 82, 85]
```

```python
# 그래프 구현

# 1행 1열의 그래픽 영역을 갖는 ax객체 생성
fig, ax = plt.subplots(1, 1)

# 막대 그래프 구현하기
# x=그룹단위, hue=그룹안에서의 구분단위, y=데이터
sns.barplot(x=subject, hue=season, y=point, palette="RdBu_r", ax=ax)

# 범주 표시하기
ax.legend()

# y축 텍스트 표시
ax.set_ylabel('캐릭터', color='#ff0000', fontsize=14)

# x축 텍스트 표시
ax.set_xlabel('득표수', color='#0000ff', fontsize=14)

# 그래프 제목 구성
ax.set_title('과목별 시험 점수', color='#ff6600', fontsize=28, pad=16)

# 배경 격자 구성
ax.grid()

# 이미지로 저장
#plt.savefig('seaborn4.png', dpi=200, bbox_inches='tight')

# 그래프 화면 출력
plt.show()

# 그래프 종료
plt.close()
```

### 산점도 그래프 구현

```python
# 샘플 데이터

# 온도와 아이스크림 판매 수량
tmp = [ 23, 25, 26, 27, 28, 29, 30, 31, 33 ]
qty = [ 21, 23, 25, 28, 33, 35, 36, 32, 39 ]
```

```python
# 산점도 그래프 구현

# 가로 18inch, 세로 10inch의 1행 1열 그래프 영역 ax 준비
fig, ax = plt.subplots(1, 1, figsize=(18, 10))

# 산점도 그래프 표시하기
sns.scatterplot(x=tmp, y=qty, label='판매수량', ax=ax)

# 범주 표시하기
ax.legend()

# 배경 격자 표시하기
ax.grid()

# 그래프 타이틀 설정
ax.set_title('기온과 아이스크림 판매수량의 관계', fontsize=24, pad=12)

# x축의 텍스트 설정
ax.set_ylabel('아이스크림 판매수량', fontsize=18, color='#0066ff')

# y축의 텍스트 설정
ax.set_xlabel('기온', fontsize=18, color='#0066ff')

# 이미지 저장
#plt.savefig("scatter.png", dpi=200, bbox_inches='tight')

# 그래프 표시하기
plt.show()

# 그래프 관련 설정 해제
plt.close()
```
