```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 샘플데이터 가져오기

교통사고xlsx = pd.read_excel("http://itpaper.co.kr/data/traffic_accident.xlsx", engine="openpyxl")
교통사고xlsx.head()
```

```python
교통사고df = 교통사고xlsx.set_index("년도")
교통사고df.head()
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 16
plt.rcParams["figure.figsize"] = (20, 10)
plt.rcParams["axes.unicode_minus"] = False
```

### 선그래프 그리기

```python
fig, ax = plt.subplots(1, 1)
교통사고df['발생건수'].plot(color='#ff0000', ax=ax)
ax.grid()
ax.legend()
ax.set_title("년도별 교통사고 발생현황", fontsize=30, pad=20)
ax.set_xlabel("년도", fontsize=20)
ax.set_ylabel("교통사고 수", fontsize=20)
plt.show()
plt.close()
```

```python
xtext = []
for v in 교통사고df.index:
    xtext.append("%d년" % v)
    
print(xtext)
```

### 파이 그래프

#### 파이 그래프 기본형


```python
최근5년df = 교통사고df.sort_index(ascending=False).head()
최근5년df
```

#### 파이 그래프에 옵션 지정하기

```python
label_list = []

for i in 최근5년df.index:
    label_list.append("%d년도" % i)

print(label_list)
```

#### 함수의 파라미터를 활용한 옵션지정

```python
<font color='wheit'>bar</font>
```

```python
fig, ax = plt.subplots(1, 1)

최근5년df['발생건수'].plot.pie(labels=label_list,  # 색션별 레이블
    colors=['#ff6600', '#0066ff', '#ffff00', "#ff00ff", "#ff3355"],  # 색션별 색상
    explode=[0.0, 0.1, 0.0, 0.2, 0.0], # 색션별 확대 비율
    autopct='%0.1f%%',  # 수치 표시 형식
    shadow=True,        # 그림자 표시 허용
    startangle=90,      # 그래프 회전 각도(반시계 방향)
    ax=ax)              # 그래프가 출력될 영역

# 제목지정
ax.set_title("교통사고 비율", fontsize=28)

# y축 라벨 표시 안함
ax.set_ylabel(None)

plt.show()
```

### 출력값에 대한 사용자 정의함수 만들기

#### 출력형식을 반환할 함수 정의하기

```python
def make_autopct(비율):
    # 함수 외부의 변수를 함수 내부에서 사용할 수 있게 하는 global 선언 사용
    global 최근5년df

    합계 = 최근5년df['발생건수'].sum()
    수치 = int(round(비율*합계/100.0))      # 비율과 총합을 사용해서 수치값을 역산한다.
    str = "%0.1f%%\n(%d건)" % (비율, 수치)  # 그래프의 각 섹션에 출력할 문자열 만들기

    return str
```

#### 사용자 정의 출력형식을 사용하여 파이 그래프 출력

```python
fig, ax = plt.subplots(1, 1)

최근5년df['발생건수'].plot.pie(labels=label_list,  # 색션별 레이블
    colors=['#ff6600', '#0066ff', '#ffff00', "#ff00ff", "#ff3355"],  # 색션별 색상
    explode=[0.0, 0.1, 0.0, 0.2, 0.0], # 색션별 확대 비율
    autopct=make_autopct,  # 수치 표시 형식
    shadow=True,           # 그림자 표시 허용
    startangle=90,         # 그래프 회전 각도(반시계 방향)
    ax=ax)                 # 그래프가 출력될 영역

# 제목지정
ax.set_title("교통사고 비율", fontsize=28)

# y축 라벨 표시 안함
ax.set_ylabel(None)

# 범주 지정
# labels -> 각 섹션별 제목
# bbox_to_anchor(x,y) -> 0,0을 기준으로 범주의 위치(2차원 좌표)
plt.legend(labels=label_list, bbox_to_anchor=(0.95, 0.625))

plt.show()
```
