### 계량적 MDS 사례

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.metrics import pairwise_distances
from sklearn.manifold import MDS
import pandas as pd
```

```python
df= pd.read_excel('http://itpaper.co.kr/data/eurodist.xlsx', engine='openpyxl').set_index('city')
df
```

```python
# 도시간 거리를 2차원 좌표로 변환

model = MDS(n_components=2, metric=True, dissimilarity='precomputed', random_state=777)
# 데이터셋에서 변환을 위한 기반 설정 수행.
fit = model.fit(df)
# 학습실행
coords = model.fit_transform(df)
coords
```

```python
# 산점도 그래프를 통해 x,y좌표 표시하기

fig, ax = plt.subplots(1, 1, figsize=(15, 15), dpi=200)
ax.scatter(coords[:,0], coords[:,1])
ax.grid()

for label, x, y in zip(df.columns, coords[:,0], coords[:,1]):
    ax.text(x, y+25, label, fontsize=12, horizontalalignment='center', verticalalignment='bottom')

# y축 좌표 0을 가로지르는 가로 직선 그리기
ax.axhline(0, ls='--', color='#ff6600')
# x축 좌표 0을 가로지르는 세로 직선 그리기
ax.axvline(0, ls='--', color='#ff6600')
    
plt.savefig('mds1.png', dpi=200, bbox_inches='tight')
plt.show()
plt.close()
```

#### 이진 변수 그룹핑

```python
df= pd.read_excel('http://itpaper.co.kr/data/head_injury.xlsx', engine='openpyxl')
df
```

```python
# 컬럼명 한글로 재설정

df2 = df.rename(columns={
    "age.65": "나이 65세 이상/미만",
    "amnesia.before": "기억상실 여부",
    "basal.skull.fracture": "골절여부",
    "GCS.decrease": "GCS 감소여부",
    "GCS.13": "GCS 초기 감소여부",
    "GCS.15.2hours": "GCS 15시간 이후 감소여부",
    "high.riks": "고위험 여부",
    "loss.of.consciousness": "의식 상실",
    "open.skull.fracture": "골절여부",
    "vomiting": "구토여부",
    "clinically.important.brain.injury": "임상적 주요 뇌손상여부"
})
df2
```

```python
# 상위 100건만 추출

df2 = df2.iloc[0:100]
print(df2.shape)
df2.head()
```

```python
# 전치 형태로 변환

df_mds = df2.T
df_mds
```

```python
# 유사성 거리 구하기
# metric : 개체간의 거리리를 계산할 때 사용할 지표 (미터와 같은 단위로 이해해도 좋음)
# jaccard : 두 개체의 총 원소 개수 중 공통 원소 개수로 유사성을 측정한다.
            # 측정된 값의 수치가 아니라 수량으로 측정해야 할 경우 사용함.
            # 두 집합에 공통으로 공유되는 항목은 중요한 반면, 두 집합에서 모두 존재하지 않는 항목에 대해서는 무시해도 되는 상황이나 문제에 적합한 비유사성 측도임
# euclidean : 두 개체간의 거리 계산에 유클리드 거리를 사용함 (기본값)
            # 측정된 값 자체가 크기나 순서를 의미하는 경우 사용함.
            # 변량의 차가 작을 수록 거리가 가깝다는 것을 의미하므로 유사도 측정에 속함.

distance = pairwise_distances(np.array(df_mds), metric='jaccard')
distance
```

```python
# 유사성 거리를 2차 평면 좌표로 변환

model = MDS(n_components=2, metric=True, dissimilarity='precomputed', random_state=777)
fit = model.fit(distance)
coords = model.fit_transform(distance)
coords
```

```python
# 좌표값을 평면 좌표값 상에 표시

plt.rcParams["font.family"] = 'Malgun Gothic' # 한글 지원 폰트 설정
plt.rcParams["font.size"] = 14                # 기본 글자크기 설정
plt.rcParams["axes.unicode_minus"] = False    # 유니코드 뺄셈기호 표시 사용 안함(깨짐방지)

fig, ax = plt.subplots(1, 1, figsize=(17, 17), dpi=200)
ax.scatter(coords[:,0], coords[:,1])
ax.grid()

for label, x, y in zip(df2.columns, coords[:,0], coords[:,1]):
    ax.text(x, y, label, fontsize=12, horizontalalignment='center', verticalalignment='bottom')

# y축 좌표 0을 가로지르는 가로 직선 그리기
ax.axhline(0, ls='--', color='#ff6600')
# x축 좌표 0을 가로지르는 세로 직선 그리기
ax.axvline(0, ls='--', color='#ff6600')
    
#plt.savefig('mds2.png', dpi=200, bbox_inches='tight')
plt.show()
plt.close()
```

### 비계량적 MDS 사례

#### 미국 50개주의 각 도시에서 발생한 사건 유형에 근거하여 유형화 하기

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.manifold import MDS
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_excel('http://itpaper.co.kr/data/us_crime.xlsx', engine='openpyxl')
df.info()
```

```python
# 결측치 확인

df.isna().sum()
```

```python
df_del = df.dropna()
df_del.isna().sum()
```

```python
# 상위 100 건만 추출 + 관련없는 변수 제거

x = df_del.iloc[0:100, 3:]
x
```

```python
# 데이터 표준화 하기

scaler = MinMaxScaler()
scaler.fit(x)
x_scaled = scaler.fit_transform(x)
#x_scaled
```

```python
# 다차원 분석 수행

model = MDS(n_components=2, metric=False, dissimilarity='euclidean', random_state=777)
model.fit(x_scaled)

coords = model.fit_transform(x_scaled)
# coords
```

```python
# 시각화 구현

plt.rcParams["font.family"] = 'Malgun Gothic' # 한글 지원 폰트 설정
plt.rcParams["font.size"] = 14                # 기본 글자크기 설정
plt.rcParams["axes.unicode_minus"] = False    # 유니코드 뺄셈기호 표시 사용 안함(깨짐방지)

fig, ax = plt.subplots(1, 1, figsize=(17, 17), dpi=200)
ax.scatter(coords[:,0], coords[:,1])
ax.grid()

for label, x, y in zip(df_del['도시이름'], coords[:,0], coords[:,1]):
    ax.text(x, y, label, fontsize=12, horizontalalignment='center', verticalalignment='bottom')

# y축 좌표 0을 가로지르는 가로 직선 그리기
ax.axhline(0, ls='--', color='#ff6600')
# x축 좌표 0을 가로지르는 세로 직선 그리기
ax.axvline(0, ls='--', color='#ff6600')
    
plt.show()
plt.close()
```
