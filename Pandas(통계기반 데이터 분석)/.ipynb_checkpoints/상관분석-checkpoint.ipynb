### 상관분석을 위한 Python 코드

```python
import numpy as np
```

#### 분산

```python
x = [1, 2, 3, 4, 5]
np.var(x)
```

#### 공분산

```python
x = [8, 3, 6, 6, 9, 4, 3, 9, 3, 4]
y = [6, 2, 4, 6, 10, 5, 1, 8, 4, 5]
np.cov(x, y)
```

```python
# 공분산 결과값만 출력

np.cov(x, y)[0, 1]
```

#### 피어슨 상관계수

```python
from scipy import stats
x = [8, 3, 6, 6, 9, 4, 3, 9, 3, 4]
y = [6, 2, 4, 6, 10, 5, 1, 8, 4, 5]
stats.pearsonr(x, y)
```

#### 스피어만 상관계수

```python
x = [8, 3, 6, 6, 9, 4, 3, 9, 3, 4]
y = [6, 2, 4, 6, 10, 5, 1, 8, 4, 5]
stats.spearmanr(x, y)
```

#### 샘플데이터 준비

```python
import numpy as np
from scipy import stats
import pandas as pd
```

```python
mtcars = pd.read_excel('http://itpaper.co.kr/data/mtcars.xlsx', engine='openpyxl')
mtcars.set_index('name', inplace=True)
mtcars
```

#### 상관분석 수행

```python
# mpg(연비)와 cyl(실린더 수)의 공분산

np.cov(mtcars['mpg'], mtcars['cyl'])
# 결과값이 음수가 나오므로 연비와 실런더 수 간에 반비례 관계가 있다고 해석
```

```python
# mpg(연비)와 cyl(실린더 수)의 상관분석
# 상관계수가 음수이므로 반비례.
# 상관계수(cor)가 `-0.852162`이므로 연비와 실린더 수는 반비례 관계이고, `1`에 가까우므로 서로 연관성이 높다고 볼 수 있다.
# 또한 p-value가 `6.113e-10`로 유의수준 `0.05`보다 작게 나타나므로 mpg와 cyl이 상관관계가 있다고 볼 수 있다.

상관계수, pvalue = stats.pearsonr(mtcars['mpg'], mtcars['cyl'])
print("상관계수: {0}, pvalue: {1}".format(상관계수, pvalue))
```

```python
# cyl(실린더 수)와 wt(무게)의 상관분석

상관계수, pvalue = stats.pearsonr(mtcars['cyl'], mtcars['wt'])
print("상관계수: {0}, pvalue: {1}".format(상관계수, pvalue))
# 상관계수가 1에 가까움으로 실린더가 많을수록 더 무거운 경향이 있다는 것을 알 수 있다.
```

#### 상관 행렬 히트맵

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (10, 10)
plt.rcParams["axes.unicode_minus"] = False
```

```python
# 히트맵으로 구성할 상관행렬
corr = mtcars.corr()
# 히트맵 그리기
ax = sns.heatmap(corr)
# 그래프 타이틀 지정
plt.title('Heatmap of MTCARS by seaborn', fontsize=16)
# 화면 표시
plt.show()
```

```python
# 히트맵을 표현할 상관행렬
corr = mtcars.corr()
# 히트맵 그리기
ax = sns.heatmap(corr,                    # 데이터
                 square=True,             # 전체 그래프를 정사각형으로 배열
                 cmap='RdYlGn',           # 색상 팔래트 이름
                 linewidths=0.1,          # 선 굵기
                 linecolor="white",       # 선 색상
                 annot=True,              # 상관계수 출력 여부
                 annot_kws={"size" : 14}) # 상관계수 글자 크기
# 그래프 타이틀
plt.title('Heatmap of MTCARS by seaborn', fontsize=16)
# 그래프 화면 표시
plt.show()
```

### 교통사고 발생건수와 부상자 수의 상관분석

```python
# 데이터셋 가져오기

traffic = pd.read_excel('http://itpaper.co.kr/data/traffic.xlsx', engine='openpyxl')
traffic.set_index(['년도','월'], inplace=True)
traffic
```

#### 교통사고 발생건수와 부상자수, 사망자수에 대한 상관분석

```python
# 부상자수에 대한 상관분석

상관계수, pvalue = stats.pearsonr(traffic['발생건수'], traffic['부상자수'])
print("상관계수: {0}, pvalue: {1}".format(상관계수, pvalue))
```

```python
# 사망자수에 대한 상관분석

상관계수, pvalue = stats.pearsonr(traffic['발생건수'], traffic['사망자수'])
print("상관계수: {0}, pvalue: {1}".format(상관계수, pvalue))
```

```python
# 상관 행렬

corr = traffic.corr()
corr
```

```python
# 히트맵을 통한 시각화 

# 히트맵 그리기
ax = sns.heatmap(corr,                    # 데이터
                 square=True,             # 전체 그래프를 정사각형으로 배열
                 cmap='PuBuGn',           # 색상 팔래트 이름
                 linewidths=0.1,          # 선 굵기
                 linecolor="white",       # 선 색상
                 annot=True,              # 상관계수 출력 여부
                 annot_kws={"size" : 14}) # 상관계수 글자 크기
# 그래프 타이틀
plt.title('교통사고 발생건수와 사망자,부상자수에 대한 상관관계', fontsize=16)
# 그래프 화면 표시
plt.show()
```

### 산점도 그래프

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 16
plt.rcParams["figure.figsize"] = (15, 8)
plt.rcParams["axes.unicode_minus"] = False
```

```python
# 교통사고 발생건수와 부상자수의 상관관계 산점도 그래프

traffic.plot.scatter(x='발생건수', y='부상자수', color='#ff6600', marker='o')
plt.grid()
plt.title("교통사고 발생건수와 부상자수의 상관관계")
plt.ylabel('부상자수')              # y축 라벨
plt.xlabel('발생건수')              # x축 라벨
plt.show()
```

```python
# 교통사고 발생건수와 사망자수의 상관관계 산점도 그래프

traffic.plot.scatter(x='발생건수', y='사망자수', color='#0066ff', marker='o')
plt.grid()
plt.title("교통사고 발생건수와 사망자수의 상관관계")
plt.ylabel('사망자수')              # y축 라벨
plt.xlabel('발생건수')              # x축 라벨
plt.show()
```

### 추세선

#### Python에서 추세선에 대한 기울기와 절편 확인하기

```python
# 추세선 방정식 구하기

x = traffic['발생건수']
y = traffic['부상자수']

# 추세선을 위한 계산 - 1차원의 polynomial(다항식)을 계산하기 위한 코드.
z = np.polyfit(x, y, 1)   # (X,Y,차원) 정의
a, b = z
print("기울기={0}, 절편={0}".format(a, b))
```

```python
z
```

```python
print("기울기={0}, 절편={1}".format(a, b))
```

#### 추세선에 대한 1차 방정식 만들기

```python
f = np.poly1d(z)
f
```

```python
f(traffic['발생건수'])
```

```python
수식 = "f(x) = %s" % str(f).strip()
수식
```

```python
# 한글폰트, 그래픽 크기 설정
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 16
plt.rcParams["figure.figsize"] = (20, 10)

traffic.plot.scatter(x='발생건수', y='부상자수', color='#ff6600', marker='o')
plt.grid()
plt.title("교통사고 발생건수와 부상자수의 상관관계")
plt.ylabel('부상자수')              # y축 라벨
plt.xlabel('발생건수')              # x축 라벨
plt.plot(x,f(x),'#ff0000')         # 추세선

plt.text(x.min()+500, y.max()-1500, 수식, fontsize=40, color='#ff0000', 
                    horizontalalignment='left', verticalalignment='top')

plt.show()
```

#### Seaborn을 활용한 산점도그래프 + 추세선

```python
# 한글폰트, 그래픽 크기 설정
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 16
plt.rcParams["figure.figsize"] = (20, 10)

# seaborn을 활용하여 산점도+추세선 일괄 표시
sns.regplot(x='발생건수', y='부상자수', data=traffic)
plt.grid()
plt.title("교통사고 발생건수와 부상자수의 상관관계")
plt.ylabel('부상자수')              # y축 라벨
plt.xlabel('발생건수')              # x축 라벨

# 수식을 표현하는 것은 개별적으로 처리해야 함
plt.text(x.min()+500, y.max()-1500, 수식, fontsize=40, color='#ff0000', 
                    horizontalalignment='left', verticalalignment='top')

plt.show()
```
