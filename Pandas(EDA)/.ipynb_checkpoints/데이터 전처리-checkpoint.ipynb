### 컬럼, 인덱스 이름 변경

```python
# 컬럼 이름 변경하기

df6 = 성적표.rename(columns={'국어':'kor', '영어':'eng', '수학':'math', '과학': 'sinc'})
df6
```

```python
# 특정 컬럼을 인덱스로 변환

df8 = 성적표.set_index('학년')
df8
```

### 데이터 타입 변경

```python
# 변수의 데이터 타입 확인

성적표.dtypes
```

```python
# 특정 컬럼을 정수로 변환

성적표['국어'] = 성적표['국어'].astype(np.int64)
```

```python
# 특정 컬럼을 실수로 변환

성적표['국어'] = 성적표['국어'].astype(np.float64)
```

```python
# 복수개의 컬럼 타입 변환

타입변환df = 성적표.astype({'국어': np.int64, '영어': np.float64})
타입변환df.dtypes
```

```python
# 정수, 실수 구분없이 숫자형 변환

성적표['국어'] = 성적표['국어'].apply(pd.to_numeric)
성적표['국어'].dtypes
```

### 범주형 변수로 변환하기

```python
# 개별컬럼 변환하기

성적표['학년'] = 성적표['학년'].astype('category')
성적표['성별'] = 성적표['성별'].astype('category')
성적표.dtypes
```

```python
# 복수 컬럼 일괄 변환

타입변환df = 성적표.astype({'학년': 'category', '성별': 'category'})
타입변환df.dtypes
```

```python
# 카테고리 값 변경하기

성적표['학년'] = 성적표['학년'].cat.rename_categories({1: '일학년', 2: '이학년', 3:'삼학년', 4:'사학년'})
성적표
```

### 데이터 정렬

```python
# 데이터 다시 호출

xlsx = pd.read_excel('http://itpaper.co.kr/data/grade_card.xlsx', 
                   engine='openpyxl')
성적표 = xlsx.set_index('이름')
```

```python
# 컬럼값 오름차순 정렬

오름차순 = 성적표.sort_values('국어')
오름차순
```

```python
# 컬럼값 내림차순 정렬

내림차순 = 성적표.sort_values('국어', ascending=False)
내림차순
```

```python
# 두 개 이상의 컬럼을 기준으로 정렬

성적표.sort_values(['국어', '수학'], inplace=True)
성적표
```

#### 인덱스를 기준 정렬

```python
# 순차 정렬

인덱스정렬1 = 성적표.sort_index()
인덱스정렬1
```

```python
# 역순 정렬 리턴없이 바로 원본에 반영

성적표.sort_index(ascending=False, inplace=True) # inplace=true가 바로 저장임
성적표
```

### 데이터 필터링 (데이터 추출)

#### 행 단위 조건 검색

```python
# 단일조건 - 국어 점수가 90점 초과인 학생

# 성적표[성적표['국어'] > 90] 이방법도있음
성적표.query('국어 > 90')
```

```python
# AND 조건 

성적표.query('국어 > 80 and 수학 > 80')
```

```python
# OR 조건

성적표.query('국어 < 70 or 수학 < 70')
```

```python
# IN 조건

# 국어점수가 item 리스트에 포함되어 있는 데이터 찾기
item = [98, 88]
# 조건문 안에 다른 변수를 끌어오기 위해서는 @변수명 형식으로 표기한다.
성적표.query('국어 in @item')
```

```python
# NOT IN 조건

# 국어점수가 item 리스트에 포함되지 않는 데이터 찾기
성적표.query('국어 not in @item')
```

```python
# 열 단위 필터링

# 이런방식도 됨 성적표[['국어', '영어']]
국어영어 = 성적표.filter(['국어','영어'])
국어영어
```
