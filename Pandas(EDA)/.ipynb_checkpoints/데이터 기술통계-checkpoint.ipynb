```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 샘플 데이터 준비

성적표 = pd.read_excel("http://itpaper.co.kr/data/grade_card.xlsx", engine='openpyxl')
성적표
```

```python
# 이름, 학년, 성별 데이터를 명목척도로 변환

성적표['이름'] = 성적표['이름'].astype('category')
성적표['학년'] = 성적표['학년'].astype('category')
성적표['성별'] = 성적표['성별'].astype('category')
성적표.dtypes
```

#### 기초 통계량 확인

```python
# 데이터 프레임 전체에 대한 확인

des = 성적표.describe()
des
```

```python
# 특정 컬럼에 대한 확인

des = 성적표.describe()
des
```

```python
# 데이터 수
print("데이터 수")
print(성적표.count())
print()
# 최대값
print("최대값")
print(성적표.max())
print()
# 표준편차
print("표준편차")
print(성적표.std())
print()
# 1사분위 수
print('1사분위 수 하위 25%위치의 값')
print(성적표.quantile(q=0.25))
print()
# 2사분위 수
print('2사분위 수, 중앙값(미디안)')
print(성적표.quantile(q=0.5))
print()
# 3사분위 수
print('3사분위 수 75%위치의 값')
print(성적표.quantile(q=0.75))
print()
# 평균
print('평균')
print(성적표.mead())
print()
# 최소값
print('최소값')
print(성적표.min())
print()
# 중앙값
print('중앙값')
print(성적표.median())
print()
# 사분위 수 응용
print('응용 - 상위 10%에 들기위한 점수')
print(성적표.quantile(q=0.9))
print()
```
