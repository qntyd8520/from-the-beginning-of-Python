### IMDB 영화 리뷰 데이터셋 (영문)

#### 패키지 준비

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

from tensorflow.keras.models import Sequential

# 학습층을 위한 패키지에 Embedding 추가
from tensorflow.keras.layers import Dense, Embedding, Flatten
from tensorflow.keras.datasets import imdb
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# 토큰화 패키지
from tensorflow.keras.preprocessing.text import Tokenizer
 # 문장 길이를 통일 시키는 용도
from tensorflow.keras.preprocessing.sequence import pad_sequences

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

#### 데이터 준비

```python
# 벡터의 최대 길이 설정 (=가장 빈번하게 나타나는 단어의 갯수)

num_words = 10000
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=num_words)
print(f'훈련 데이터 {x_train.shape} 레이블 {y_train.shape}')
print(f'검증 데이터 {x_test.shape} 레이블 {y_test.shape}')
```

#### 데이터 전처리

```python
# 전처리가 완료된 데이터 셋임으로 생략
```

#### EDA

```python
# 첫 번째 데이터 확인하기

print(x_train[0])
```

```python
# 첫 번째 레이블값 확인

print(y_train[0])
```

```python
# 각 단어별 토큰번호 확인하기

word_index = imdb.get_word_index()
word_index
```

```python
# 임의의 훈련 데이터에 대해 원 단어를 대입하여 문장 구성하기

# 문장의 각 단어가 저장될 빈 리스트
decoded_review = []

# 첫 번째 데이터의 각 단어를 의미하는 토큰에 대한 반복
for i in x_train[0]:
    # 단어별 토큰에서 찾은 실제 단어가 저장될 변수
    find = "?"
    
    # 단어별 토큰번호 딕셔너리에서 반복
    for word in word_index:
        # 단어에 대한 토큰번호 취득
        index = word_index[word]
        # 첫 번째 데이터의 토큰번호와 일치하는 토큰번호가 있다면
        if i-3 == index:
            # 검출된 토큰번호에 맵핑된 단어를 추출
            find = word
            break
    
    # 추출된 단어를 빈 리스트에 저장
    decoded_review.append(find)

# 단어가 모인 리스트를 공백을 기준으로 병합
review_text = ' '.join(decoded_review)
review_text
```

```python
# 데이터를 동일한 길이로 맞추기

# 최대 문장 길이
max_len = 500

print('Before pad_sequences: ', len(x_train[0]))

pad_x_train = pad_sequences(x_train, maxlen=max_len, padding = 'pre')
pad_x_test = pad_sequences(x_test, maxlen=max_len, padding = 'pre')

#  원래 단어의 앞에 ‘지정해준 단어의 길이 – 원래 단어의 길이’(500 –218)만큼 0이 추가된 것을 볼 수 있다.
print('After pad_sequences: ', len(pad_x_train[0]))

print(pad_x_train[0])
```

#### 데이서셋 분할

```python
# 랜덤시드 고정

np.random.seed(777)
```

#### 모델 개발

```python
# 모델 정의

model = Sequential()

# Embedding층은 모델의 첫 번째 층으로만 사용할 수 있으며, 주로 순환 신경망과 연결하여 사용한다.
# 1) input_dim 인자는 학습 데이터셋에서 사용한 단어의 개수(10,000)
# 2) output_dim 인자는 임베딩 벡터의 크기(32)
# -> 이와 같이 인자를 설정하고 나면 Embedding층은 (None, 500, 32) 형태의 출력을 가지게 됩니다.
# -> 500은 입력값의 길이를 지정했던 max_len의 길이와 같으며, 32는 임베딩 벡터의 크기와 같습니다.
# 3) input_length인자는 Flatten 층을 사용하기 위해 전달.
model.add(Embedding(input_dim = num_words, output_dim = 32, input_length = max_len))

# Embedding층 이후에 Dense층을 연결
model.add(Flatten())
model.add(Dense(1, activation = 'sigmoid'))

# 결국은 문장을 학습하여 긍정, 부정 여부를 판단하는 것이므로 이진분류분석에 속한다.
model.compile(optimizer='adam',loss = 'binary_crossentropy', metrics = ['acc'])
model.summary()
```

```python
# 학습하기

result = model.fit(pad_x_train, y_train, batch_size = 32, epochs = 30, validation_data=(pad_x_test, y_test), callbacks = [
    EarlyStopping(monitor = 'val_loss', patience=5, verbose = 1),
    ReduceLROnPlateau(monitor= "val_loss", patience=3, factor = 0.5, min_lr=0.0001, verbose=1)
])

result_df = pd.DataFrame(result.history)
result_df['epochs'] = result_df.index+1
result_df.set_index('epochs', inplace=True)
result_df
```

#### 학습 결과 평가

```python
# 학습 결과 시각화

# 그래프 기본 설정
# ----------------------------------------
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 16
plt.rcParams['axes.unicode_minus'] = False

# 그래프를 그리기 위한 객체 생성
# ----------------------------------------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5), dpi=150)

# 1) 훈련 및 검증 손실 그리기
# ----------------------------------------
sns.lineplot(x=result_df.index, y='loss', data=result_df, color='blue', label='훈련 손실률', ax=ax1)
sns.lineplot(x=result_df.index, y='val_loss', data=result_df, color='orange', label='검증 손실률', ax=ax1)
ax1.set_title('훈련 및 검증 손실률')
ax1.set_xlabel('반복회차')
ax1.set_ylabel('손실률')
ax1.grid()
ax1.legend()

# 2) 훈련 및 검증 정확도 그리기
# ----------------------------------------
sns.lineplot(x=result_df.index, y='acc', data=result_df, color = 'blue', label = '훈련 정확도', ax=ax2)
sns.lineplot(x=result_df.index, y='val_acc', data=result_df, color = 'orange', label = '검증 정확도', ax=ax2)
ax2.set_title('훈련 및 검증 정확도')
ax2.set_xlabel('반복회차')
ax2.set_ylabel('정확도')
ax2.grid()
ax2.legend()
```

```python
# 훈련데이터에 대한 모델 성능 평가

evaluate = model.evaluate(pad_x_train, y_train)
print("훈련 데이터 최종 손실률: %f, 최종 정확도: %f" % (evaluate[0], evaluate[1]))
```

```python
# 검증 데이터에 대한 모델 성능 평가

evaluate = model.evaluate(pad_x_test, y_test)
print("검증 데이터 최종 손실률: %f, 최종 정확도: %f" % (evaluate[0], evaluate[1]))
```

#### 학습 결과 적용

```python
# 테스트 데이터에 대한 예측 결과 산정

result = model.predict(pad_x_test)
data_count, case_count = result.shape
print("%d개의 검증 데이터가 %d개의 경우의 수를 갖는다." % (data_count, case_count))
result
```

```python
# 예측 결과를 1차원 배열로 변환

f_results = result.flatten()
f_results
```

```python
# 결과 데이터셋 구성

kdf = pd.DataFrame({
    '결과값': y_test,
    '예측치' : np.round(f_results)
})

kdf['예측치'] = kdf['예측치'].astype('int')

kdf['결과값'] = kdf['결과값'].replace({0: '긍정', 1:'부정'})
kdf['예측치'] = kdf['예측치'].replace({0: '긍정', 1:'부정'})

kdf
```

```python
# 실제 결과값과 머신러닝에 의한 예측값 비교

# 예측 데이터(ca_y_test)를 x축으로, 결과 데이터(results)를 y축으로 놓는 오차행렬(confusion_matrix) 생성하기 
cm = confusion_matrix(kdf['결과값'], kdf['예측치'])
cmdf1 = pd.DataFrame(cm, columns=['예측값(P)', '예측값(N)'], index=['실제값(T)', '실제값(F)'])
cmdf1
```

```python
# 오차행렬 히트맵

sns.heatmap(cm, annot = True, fmt = 'd',cmap = 'Blues')
plt.xlabel('예측값')
plt.ylabel('결과값')
```
